### 인덱싱 적용 방법

|적용방법|구문|결과타입|
|:---|:---:|:---:|
|열선택|df[col]|Series|
|라벨로 행 선택|df.loc[label]|Series|
|정수 위치로 행 선택|df.iloc[loc]|Series|
|행 슬라이스|df[5:10]|DataFrame|
|불리언 벡터로 행 선택|df[bool_vec]|DataFrame|

````python
import pandas as pd
import seaborn as sns
import numpy as np
````

````python
friend_dict_list = [
    {
        "name": "John",
        "age": 25,
        "job": "student",
    },
    {
        "name": "Nate",
        "age": 34,
        "job": "teacher",
    },
    {
        "name": "Jenny",
        "age": 30,
        "job": "developer",
    },
]

friend_df=pd.DataFrame(friend_dict_list)
friend_df
````

### rows, columns 기준

#### loc[]
- 5 또는 a 와 같은 단일 라벨
- ['a','b','c']와 같은 리스트나 라벨의 배열
- a:f 와 같이 라벨이 있는 처음과 끝의 범위를 포함하는 슬라이스 객체
- [True, False, True]와 같이 슬라이싱할 축과 같은 길이의 불리언 배열
- 호출하는 시리즈나 데이터 프레임을 인수로 가지는 호출 함수

#### iloc[]
- 정수
- [4,3,0] 과 같은 리스트나 정수들의 배열
- 1:7 과 같은 정숫값을 가지는 슬라이스 객체
- 불리언 배열
- 호출하는 시리즈나 데이터프레임을 인수로 가지는 호출 함수

````python
friend_df.loc[0] # 인덱스 0번 데이터를 조회함 이렇게 조회하면 시리즈로 반환함
````

````python
friend_df.iloc[0:1] # 이렇게 iloc 슬라이싱을 쓰면 데이터 프레임으로 반환하지만 1번은 포함하지 않는다
````

````python
friend_df.iloc[0:3]
````

````python
# 기본 데이터 셋 가져오기

iris_df=sns.load_dataset("iris")
iris_df
````

````python
iris_df.iloc[10:20] # 10 ~ 19
````

><b>df.iloc[ ] : 행, 열 모두 position 값을 이용(integer 값)</b>

````python
iris_df.iloc[1:3][["sepal_width","petal_length"]] # 이 방법도 있고
iris_df.iloc[1:3,1:3] # 이렇게 인덱스 기준으로 할 수도 있다
````

````python
# 행 index는 모두 포함, 열 index는 1, 3번 컬럼만 포함
iris_df.columns


iris_df.iloc[:,[1,2,4]]
````

><b>df.loc[ ] : 행, 열의 label 값을 이용하거나, boolean 이용 / 마지막 값 포함</b>

````python
iris_df.loc[1]
````

````python
iris_df.loc[11:20]
````

````python
 iris_df.loc[1,'sepal_width':'petal_width'] # 문자열로 집어놓은 것도 범위로 지정 가능
````

````python
iris_df.loc[20:30][["sepal_width"."petal_length","petal_width"]]
````

````python
iris_df[1:3][["sepal_width","petal_width"]]
````

````python
# iris_df[0:149][["sepal_length","sepal_width","petal_length","petal_width","species"]]

# sepal_width 이 3.0 보다 큰 데이터만 추출

iris_df.loc[iris_df["sepal_width"] > 3.0]
````

````python
# sepal_width 이 3.0 보다 큰 데이터만 추출이지만 2개의 컬럼만 보고싶다

iris_df.loc[iris_df["sepal_width"] > 3.0][["sepal_width","petal_width"]]
````

#### 3) filter(items=None, like=None, regex=None, axis=None)
```
행 또는 열의 부분 집합 지정
```

````python
# _가 들어있는 컬럼 가져오기
iris_df.filter(regex="_")
````

````python
# length 로 끝나는 컬럼 가져오기
iris_df.filter(regex="length$") # $ : 끝나는

````

````python
# Sepal 로 시작하는 컬럼 가져오기
iris_df.filter(regex="^sepal")

````

````python
# se로 시작하는 컬럼
iris_df.filter(regex="^se") # ^ 시작하는

````

````python
# u_list={"one":[1,4],"two":[2,5],"three":[3,6]}
# df=pd.DataFrame(u_list,index=["mouse","rabbit"])
# df 이 방법도 있지만

# 바로 주는 방법도 있다
df=pd.DataFrame(data=[[1,2,3],[4,5,6]],index=["mouse","rabbit"],columns=["one","two","three"])
df

````

````python
df.filter(items=["one","three"])
````

````python
df.filter(regex="e$",axis=0)
````

````python
df.filter(like="bbi",axis=0)
````
